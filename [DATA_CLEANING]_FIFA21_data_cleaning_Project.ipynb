# **DATA CLEANING**

Create a copy of dataframe

```python
 fifa = data.copy()
 fifa.sample(20)
```

1. Club Column (current club of the player)

```python
fifa['Club'].dtype
```

```python
fifa['Club'].unique()
```

```python
fifa['Club'] = fifa['Club'].str.strip() #to removing leading and trailing whitespace,including newline characters ('\n')
fifa['Club'].unique()
```

2. Contract (player contract details)

```python
fifa['Contract'].dtype
```

```python
fifa['Contract'].unique()
```

```python
#see the same format and same case, consitent

for index, row in fifa.iterrows():
    if 'On Loan' in row['Contract'] or 'Free' in row['Contract']:
        print(row['Contract'])
```

```python
#function to extract data from 'Contract' column
def extract_contract_info(contract):
    if contract == 'Free' or 'On Loan' in contract:
        start_date = np.nan
        end_date = np.nan
        contract_lenght = 0
    else:
        start_date, end_date = contract.split(' ~ ')
        start_year = int(start_date[:4])
        end_year = int(end_date[:4])
        contract_lenght = end_year - start_year
    return start_date, end_date, contract_lenght

#apply fn to 'Contract' column and create new columns

new_cols = ['Contract_Start', 'Contract_End', 'Contract Lenght(years)']
new_data = fifa['Contract'].apply(lambda x: pd.Series(extract_contract_info(x)))

for i in range(len(new_cols)):
    fifa.insert(loc=fifa.columns.get_loc('Contract')+1+i, column=new_cols[i], value=new_data[i])
 
```

```python
fifa.sample(10)
```

```python
#contract categories

def categories_contract_status(contract):
    if contract == 'Free':
      return 'Free'
    elif 'On Loan' in contract:
      return 'On Loan'
    else:
      return 'Contract'

#add contract status column

fifa.insert(fifa.columns.get_loc('Contract Lenght(years)')+1, 'Contract Status', fifa['Contract'].apply(categories_contract_status))
```

```python
fifa.head()
```

```python
#view a subset of our new data

fifa[['Contract', 'Contract_Start', 'Contract_End', 'Contract Lenght(years)', 'Contract Status']].sample(10)
```

3. Height (Player's height in centimeters)

```python
fifa['Height'].dtype
```

```python
fifa['Height'].unique()
```

```python
def convert_height(height):
    if "cm" in height:
        return int(height.strip("cm"))
    else:
        feet, inches = height.split("'")
        total_inches = int(feet)*12 + int(inches.strip('"'))
        return round(total_inches*2.54)

# Apply the conversion function to the 'Height' column

fifa['Height'] = fifa['Height'].apply(convert_height)
fifa['Height'].unique()
```

```python
#rename the Height column to Height(cm)

fifa = fifa.rename(columns={'Height':'Height(cm)'})
fifa.sample(1)
```

4.Weight (Player's weight in kilograms)

```python
fifa['Weight'].dtype
```

```python
fifa['Weight'].unique()
```

```python
def convert_weight(weight):
    if "kg" in weight:
        return int(weight.strip("kg"))
    else:
        pounds = int(weight.strip("lbs"))
        return round(pounds/2.205)

#apply fn to Weight column
fifa['Weight'] = fifa['Weight'].apply(convert_weight)
fifa['Weight'].unique()
```

```python
#rename the Weight column
fifa = fifa.rename(columns={"Weight":"Weight(kg)"})
fifa.sample(3)
```
